```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline

data_xvals = numpy.atleast_2d( [ 1.0, 3.0, 5.0, 6.0, 7.0, 8.0 ] )
data_yvals = numpy.sin( data_xvals )

test_xvals = numpy.arange(-2, 10, 0.1)

def linear(a, b):
    return numpy.multiply(a, b)
def linearNoise(a, b):
    return linear(a, b) + delta(a, b)

def polynomial(a, b):
    return numpy.power(1 + numpy.multiply(a, b), 3)
def polynomialNoise(a, b):
    return polynomial(a, b) + delta(a, b)

def rbf(a, b):
    return numpy.exp(-0.5 * numpy.power(numpy.subtract(a, b), 2))
def rbfNoise(a, b):
    return rbf(a, b) + delta(a, b)

def delta(a, b):
    d = numpy.divide(a, b)
    c = numpy.where(d == 1, numpy.ones(d.shape) * 0.1, numpy.zeros(d.shape))
    return c

def plot(title, y_est, y1, y2):
    ax = plt.subplot()
    ax.set_title(title)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    plt.gca().fill_between(test_xvals, y1, y2, color='y')
    plt.plot(test_xvals, y_est, color='c')
    plt.scatter(data_xvals, data_yvals, color='b')
    plt.show()
    
def Gaussian(kernel, graph_title):
    mat = numpy.linalg.pinv(kernel(data_xvals, numpy.transpose(data_xvals)))
    y = list(map(lambda x: float(numpy.dot(numpy.dot(kernel(x, data_xvals), mat),
                                 numpy.transpose(data_yvals))), test_xvals))
    covar = list(map(lambda x: float(kernel(x, x) - numpy.dot(numpy.dot(kernel(x, data_xvals), mat), 
                                     numpy.transpose(kernel(data_xvals, x)))), test_xvals))
    plot(graph_title, y[:], numpy.subtract(y, covar), numpy.add(y, covar))    
```

Linear kernel:

```python
Gaussian(linear, 'Noise-free GPR, linear kernel')
Gaussian(linearNoise, 'Noisy GPR, linear kernel')
```

```python
Gaussian(polynomial, 'Noise-free GPR, polynomial kernel')
Gaussian(polynomialNoise, 'Noisy GPR, polynomial kernel')
```

When the degree M of the polynomial kernel gets small, the data is underfit and inaccurate in the learning data range, but when it gets large, it becomes overfit and inaccurate anywhere outside of learning data range.

```python
Gaussian(rbf, 'Noise-free GPR, RBF kernel')
Gaussian(rbfNoise, 'Noisy GPR, RBF kernel')
```

The bandwidth parameter is inversely proportionally related to the variance of the Gaussian. When the bandwidth decreases, the variance increases; when the bandwidth increases, the variance narrows.
