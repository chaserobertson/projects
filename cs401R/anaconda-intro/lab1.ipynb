```python
import cairo
import numpy
 
# this tells seaborn and matplotlib to generate plots inline in the notebook
%matplotlib inline  
 
# these two lines allow you to control the figure size
%pylab inline
pylab.rcParams['figure.figsize'] = (16.0, 8.0)
    
# A simple function to display an image in an ipython notebook
def nbimage( data ):
    from IPython.display import display, Image
    from PIL.Image import fromarray
    from io import BytesIO
 
    s = BytesIO()
    fromarray( data ).save( s, 'png' )
    display( Image( s.getvalue() ) )
 
WIDTH = 512
HEIGHT = 288
 
# this is a numpy buffer to hold the image data
data = numpy.zeros( (HEIGHT,WIDTH,4), dtype=numpy.uint8 )
 
# this creates a cairo context based on the numpy buffer
ims = cairo.ImageSurface.create_for_data( data, cairo.FORMAT_ARGB32, WIDTH, HEIGHT )
cr = cairo.Context( ims )
 
for i in range(0,50):
    # draw a random line of random color
    rgb = numpy.random.rand(4)
    coords = numpy.random.rand(8)

    cr.set_source_rgba( rgb[0], rgb[1], rgb[2], rgb[3] )
    cr.set_line_width( 20.0 )
    cr.move_to( coords[0] * WIDTH, coords[1] * HEIGHT )
    cr.line_to( coords[2] * WIDTH, coords[3] * HEIGHT )
    cr.line_to( coords[4] * WIDTH, coords[5] * HEIGHT )
    cr.line_to( coords[6] * WIDTH, coords[7] * HEIGHT )
    cr.stroke()
 
# display the image
nbimage( data )
```
