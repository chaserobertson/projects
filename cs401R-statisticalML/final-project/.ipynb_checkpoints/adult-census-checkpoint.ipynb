## Part 1 - The Dataset
The US census data came from the BYU CS 478 dataset collection at axon.cs.byu.edu/data/uci_class/adult-census.arff, which was extracted from US census data at www.census.gov. This data is interesting because it is anonymous, public census information that can be used to analyze the important factors in Americans' income levels.

```python
from scipy.io import arff
import pandas as pd

filename = 'datasets/adult-census.arff'

data, meta = arff.loadarff(filename)
meta_map = dict(zip(meta.names(), meta.types()))
df = pd.DataFrame(data)

df.head(3)
```

## Part 2 - The Problem
I chose to use this data to try to predict a person's income level (above or below 50k, which is near the national median income). This is a supervised classification problem: each training instance includes the income level of the instance, and classification of test instances is the goal. Many other successful attempts have been made to classify income bins based on census data; this project's goal is to explore the accuracy of different statistical models on this particular problem.

## Part 3 - The Dataset
A preliminary look at the data is shown below. The data includes features for a person's age, working class, a Census Bureau localization weight, formal education level, years of formal education, marital status, occupation, relationship, race, sex, annual capital gains, annual capital loss, hours worked per week, native country, and annual income class. I chose to visualize the numeric features by density plot, correlation matrix, and scatter matrix to try to get a feel for the usefulness of each numeric feature.

```python
cols = list(df.keys())
numeric_cols = np.array(cols)[np.where(np.array(meta.types()) == 'numeric')]
numeric_cols = list(numeric_cols)
numeric_cols.append('class >50k')
df_encoded = pd.get_dummies(df, columns=['class'], drop_first=True)
```

```python
pd.plotting.scatter_matrix(df)
plt.show()
```

After plotting each numeric feature, I found that some features were more useful than others. For example, capital gains and capital losses are inversely correlated as a result of being two measures of the same information, so I combined them below into a single net capital gain feature. The final weight feature is a Census Bureau addition that is meant to indicate geographic location, but evidently does not show much correlation with income classification.

```python
if 'capital-loss' in df.keys():
    df['capital-gain'] -= df['capital-loss']
    df.drop('capital-loss', axis=1, inplace=True)
    cols.remove('capital-loss')
    del(meta_map['capital-loss'])
df.head(0)
```

In order to visualize the nominal features of the data, I created a bar chart for each showing the distribution of possible values, all of which are shown below. I couldn't glean much from the first few feature charts, but the distribution of races was surprising to me. I was under the impression that only a slight majority of Americans were white. After further investigation, I realized that there was no "ethnicity" feature, which by US government definition would distinguish between whites and Hispanics or Latinos. Adding ethnicity information in the future could be a factor in increasing prediction accuracy.

By inspection, sex, relationship and marital-status contain some redundant information that can be condensed. For example, the 'husband' and 'wife' classes can be condensed to one 'spouse' class, with the distinction between sexes reserved for the sex feature.

The native-country feature is severely condensed in the United States class. A further exploration of this and other less obviously useful features is shown in Part 5.

```python
from collections import Counter

nominal_cols = [col for col in cols if meta_map[col] == 'nominal']
nominal_cols.remove('class')
nominal_features = df[nominal_cols]
fig, axs = plt.subplots(8, 1, figsize=(15, 25), sharex=False)
for i in range(len(nominal_cols)):
    col = nominal_cols[i]
    counter = Counter(nominal_features[col].values)
    axs[i].bar(counter.keys(), counter.values())
    axs[i].set_title(col)
plt.tight_layout()
```

## Part 4 - My Approach
My choice of approach was to contrast several statistical models on this data, revisiting the dataset to try to optimize results. I chose to use the excellent algorithms implemented by scikit-learn, so that I could try multiple models without importing multiple packages. The models that I chose to use are listed below:

1. Logistic Regression: Predictions are made based on the log probability of belonging to each output class. A linear model is learned to best fit the training data, then instances are classified by their log odds. I chose this model because it is a good linear baseline to compare the other models against.
2. Naïve Bayes: Assuming all features are independent variables, predictions are made based on the relative frequency of each feature being associated with each label in the training set. This is an extremely fast and relatively effective classifier, with a long history of use in spam filters and document classification. I chose this because it is a simple, fast model which can accurately classify simple linear associations like the data at hand.
3. Linear Discriminant Analysis: Predictions are made based on a linear decision surface which is trained by projecting the training features onto a linear subspace that maximizes the separation of classes. I chose this model because it is fast and requires no hyperparameter tuning.

Out of curiosity I also tested Quadratic Discriminant Analysis on this data, but it did not fit as well as its Linear counterpart, so I omitted that experiment.

I partitioned the data into a random 20-80% test-training split using sklearn's built-in function.

```python
from sklearn.model_selection import train_test_split

if 'class' in cols:
    cols.remove('class')
features = df[cols]
labels = df[['class']]

features_encoded = pd.get_dummies(features, columns=nominal_cols)
labels_encoded = pd.get_dummies(labels, drop_first=True)
X_train, X_test, Y_train, Y_test = train_test_split(features_encoded, labels_encoded.values[:,0], test_size=0.2, train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
lda_model = LogisticRegression()
lda_model.fit(X_train, Y_train)
print(lda_model.score(X_train, Y_train))
print(lda_model.score(X_test, Y_test))
```

```python
from sklearn import naive_bayes
gnb_model = naive_bayes.GaussianNB()
gnb_model.fit(X_train, Y_train)
print(gnb_model.score(X_train, Y_train))
print(gnb_model.score(X_test, Y_test))
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda_model = LinearDiscriminantAnalysis()
lda_model.fit(X_train, Y_train)
print(lda_model.score(X_train, Y_train))
print(lda_model.score(X_test, Y_test))
```

## Part 5 - An Analysis

My final classification accuracy on the test set (random 20% of all data) was 84%. None of the three models overfit on the training set, evidenced by the similar accuracy on the training set and the test set. I did not end up experimenting as much as I wanted to with feature reduction (family relations, native country, Census final weight, LDA collinearity), simply for lack of time. I think I made good progress on solving this problem, though I would have liked to see better results. 

If I were to continue with this problem, or recommend steps to other researchers, I would definitely add an ethnicity feature or include it in the race feature to increase variance. I would also search for unbinned income data and pivot to a regression problem, predicting exact income rather than simply above/below 50k.
