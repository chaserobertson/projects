```python
import lab8_common as util
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt

mat = loadmat('sensors.mat')
data = mat['sonars']
true_states = mat['true_states']

T = data.shape[1]
N = 500

def lap_lik( data, part_t ):
    sigma = 25
    out = np.exp(-sigma * np.sum(np.abs(data - part_t), axis=0))
    return out

def addNoise( parts ):
    sigma = 0.02
    parts[0] = parts[0] + [np.random.normal(scale=sigma) for n in range(N)]
    parts[1] = parts[1] + [np.random.normal(scale=sigma) for n in range(N)]
    parts[2] = parts[2] + [np.random.rand() * 2 * np.pi for n in range(N)]
    return parts

def show_map( room_map ):
    for i in range( 0, room_map.shape[0] ):
        plt.plot( [room_map[i,0], room_map[i,2]], [room_map[i,1], room_map[i,3]] )
    plt.xlim( -0.1, 1.1 )
    plt.ylim( -0.1, 1.1 )

def draw(particles, true_values, predicted_values):
    clear_output(wait=True)
    show_map( room_map )
    plt.scatter(particles[0,:] , particles[1,:], c="yellow")
    plt.plot(true_values[0,:],  true_values[1,:] , c= "green", label="Actual Position")
    plt.plot(predicted_values[0,:], predicted_values[1,:], c= "blue", label="Estimated Position")
    plt.legend(loc=1)
    plt.show()
    
room_map = util.create_map()
exs = np.zeros((3, T))
parts = np.zeros((3, N))
parts[0] = np.random.uniform(0, 1, N)
parts[1] = np.random.uniform(0, 1, N)
parts[2] = np.random.random_sample(N) * 2 * np.pi

for t in range(T):
    parts = addNoise(parts)
    # compute weights
    part_sensor_data = util.cast_rays( parts, room_map )
    liks = lap_lik( data[:,t:t+1], part_sensor_data.T ).T
    # normalize weights
    liks = liks / np.sum(liks)
    # resample particles
    inds = np.random.choice(range(N), size=N, p=liks)
    parts = parts[:, inds]
    exs[:, t] = np.sum(parts * liks, axis=1)
    draw(parts, true_states[:,0:t], exs[:,0:t])
```
