## Part 1: Dataset Exploration

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn
import pandas as pd

def RMSE(predictions, actuals):
    if len(predictions) == len(actuals):
        return np.sqrt(np.mean(np.square(predictions - actuals)))

ur = pd.read_csv('user_ratedmovies_train.dat','\t')
N = ur.shape[0] #total instances
M = ur.shape[1] #total features
print(N, M)
ur.head(5)
```

```python
# visualize feature distributions

for key in ur.keys():
    ax = plt.subplot(title=key)
    ax.hist( ur[key] )
    plt.show()
```

```python
# explore other available data

data_files = ['movie_countries', 'movie_genres', 'movie_tags', 'user_taggedmovies']
urs = {}
for file in data_files:
    temp = pd.read_table(file+'.dat')
    for key in temp.keys():
        ax = plt.subplot(title=file+','+key)
        ax.hist( temp[key] )
        plt.show()
```

## Part 2: Make Predictions

```python
# create a test/train split
 
all_inds = np.random.permutation( range(0,len(ur)) )
test_inds = all_inds[0:85000]
train_inds = all_inds[85000:len(ur)]
 
ur_test = ur.iloc[ test_inds ]
ur_train = ur.iloc[ train_inds ]
```

```python
# make default(mean) predictions

train_predictions = np.zeros(len(ur_train))
mean_rating = np.mean(ur_train['rating'])
for i in range(len(train_predictions)):
    train_predictions[i] = mean_rating
    
train_actuals = ur_train['rating']
train_error = RMSE(train_predictions, train_actuals)
print('Train set RMSE: ', train_error)

test_predictions = np.zeros(len(ur_test))
for i in range(len(test_predictions)):
    test_predictions[i] = mean_rating
    
test_actuals = ur_test['rating']
test_error = RMSE(test_predictions, test_actuals)
print('Test set RMSE: ', test_error)
```

```python
# train real model

def predict(u, m):
    return np.dot(p[u], q[m])

def get_error():
    predictions = np.zeros(len(ur_test))
    for i in range(len(predictions)):
        inst = ur_test.iloc[i]
        u = int(inst['userID'])
        m = int(inst['movieID'])
        predictions[i] = predict(u, m)
    return RMSE(predictions, test_actuals)

n_users = 80000 # number of users
n_movies = 80000 # number of movies
n_factors = 3  # number of factors
n_epochs = 10  # number of iterations
eta = .01  # learning rate
l = .05    # regularization

# randomly initialize user and movie factors
p = np.random.normal(0, .1, (n_users, n_factors))
q = np.random.normal(0, .1, (n_movies, n_factors))

for e in range(n_epochs):
    for x in range(len(ur_train)):
        inst = ur_train.iloc[x]
        u = int(inst['userID'])
        i = int(inst['movieID'])
        r = inst['rating']
        err = r - predict(u, i)
        # update
        p[u] = (1 - l*eta)*p[u] + eta*q[i]*err
        q[i] = (1 - l*eta)*q[i] + eta*p[u]*err

    print('Epoch %d RMSE: ' % e, get_error())
```

```python
# final prediction

predictions = np.zeros(len(ur_test))
for i in range(len(predictions)):
    inst = ur_test.iloc[i]
    u = int(inst['userID'])
    m = int(inst['movieID'])
    predictions[i] = predict(u, m)
    
test_error = RMSE(predictions, test_actuals)
print('Final prediction RMSE: ',test_error)
```

```python
# write out predictions file

pred_array = pd.read_table('predictions.dat')
test_ids = pred_array[["testID"]]
test_users = pred_array[["userID"]]
test_movies = pred_array[["movieID"]]
pred_array.head()

N = pred_array.shape[0]
my_preds = np.zeros((N,1))

for id in range(N):
    user = test_users.iloc[id][0]
    movie = test_movies.iloc[id][0]
    my_preds[ id, 0 ] = predict(user, movie)

sfile = open( 'predictions.csv', 'w' )
sfile.write( '"testID","predicted_rating"\n' )
for id in range( 0, N ):
    sfile.write( '%d,%.2f\n' % (test_ids.iloc[id], my_preds[id] ) ) 
sfile.close()
```
